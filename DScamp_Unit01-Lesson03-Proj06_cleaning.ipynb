```python
import numpy as np
import pandas as pd
```

```python
# Function : strip off special character from a string
#
def remove_special_char(s, c='\n'):
    result = ''
    tokens = str(s).split(c)        
    return result.join(tokens)
```

```python
def series_remove_special_char(series, c='\n'):
    nsize = series.size
    for i in range(nsize):
        series[i] = remove_special_char(series[i], c)
        
    return series
```

```python
# Function : remove spaces
#            at front or ar end of a string
#
def remove_space_front_end(x):
    return (str(x).lstrip()).rstrip()
```

```python
def convert_dollar_to_pound(x):
    currencyExchange = 1.25
    return (x / currencyExchange)
```

```python
df_journal = pd.read_csv('../data-101/WELLCOME_APCspend2013_forThinkful_clean01.csv', encoding='cp437')
```

```python
df_journal.head(3)
```

```python
[rows, cols] = df_journal.shape
```

   cleaning data
        1) remove spaces at front or at end of a string
        2) remove special character

```python
for i in range(rows):
    (df_journal.iloc[i]).apply(lambda x: remove_space_front_end(x))
    temp = series_remove_special_char(df_journal.iloc[i], '\n')
    df_journal.iloc[i] = temp
```

```python
print('drop the NA columns 5, 6, 7')
df_journal.drop(df_journal.columns[7], axis=1, inplace=True)
df_journal.drop(df_journal.columns[6], axis=1, inplace=True)
df_journal.drop(df_journal.columns[5], axis=1, inplace=True)
```

## create new column of cost in pound

```python
lst = [0.0]*rows
df_journal['cost_in_pound'] = lst
```

```python
for i in range(rows):                          
    if not str(df_journal.iat[i, 4][0:1]).isdigit():
        lst[i] = float(df_journal.iat[i, 4][1:])
    elif not str(df_journal.iat[i, 4])[-1:].isdigit():
        lst[i] = convert_dollar_to_pound(float(df_journal.iat[i, 4][:-1])) 
        
df_journal['cost_in_pound'] = lst
```
