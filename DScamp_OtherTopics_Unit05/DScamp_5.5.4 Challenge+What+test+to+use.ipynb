```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# import data
df = pd.read_csv('../data-201-resources-master/ESS_practice_data/ESSdata_thinkful.csv')
df.head(2)
```

## 1.  Did people become less trusting from 2012 to 2014? Compute results for each country in the sample.

```python
graph = sns.FacetGrid(df, row='cntry', col='year')
graph.map(plt.hist, "ppltrst", color="steelblue", lw=0)
plt.suptitle('People Trustiness')
```

```python
print('CH 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='CH')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='CH')]['ppltrst'].std())) 
print('CH 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='CH')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='CH')]['ppltrst'].std())) 
print('CZ 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='CZ')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='CZ')]['ppltrst'].std())) 
print('CZ 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='CZ')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='CZ')]['ppltrst'].std())) 
print('DE 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='DE')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='DE')]['ppltrst'].std())) 
print('DE 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='DE')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='DE')]['ppltrst'].std())) 
print('ES 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='ES')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='ES')]['ppltrst'].std())) 
print('ES 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='ES')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='ES')]['ppltrst'].std())) 

print('NO 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='NO')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='NO')]['ppltrst'].std())) 
print('NO 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='NO')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='NO')]['ppltrst'].std())) 

print('SE 2012 mean = {}, std = {}'.format(df[(df.year==6) & (df.cntry=='SE')]['ppltrst'].mean(),df[(df.year==6) & (df.cntry=='SE')]['ppltrst'].std())) 
print('SE 2014 mean = {}, std = {}'.format(df[(df.year==7) & (df.cntry=='SE')]['ppltrst'].mean(),df[(df.year==7) & (df.cntry=='SE')]['ppltrst'].std())) 
```

## answer:     in country CH,   people gets more trusting
                 in country CZ, people gets more trusting
                 in country DE, people gets more trusting
                 in country ES, people gets less trusting
                 in country NO, people gets less trusting
                 in country SE, people gets more trusting
